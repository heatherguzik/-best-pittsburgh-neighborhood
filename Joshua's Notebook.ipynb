## Dataset #1

```python
import pandas as pd

# Reading the csv file, displaying first ten values to test it out
arrest_data = pd.read_csv("City of Pittsburgh Arrest Data Set.csv")

arrest_data.head(10)
```

```python
# Count the number of arrests per neighborhood
arrest_counts = arrest_data["INCIDENTNEIGHBORHOOD"].value_counts().sort_values(ascending=True)

# Display the neighborhoods with the least amount of arrests
arrest_counts.head(30)
```

According to this, Mt. Olvier Neighborhood has the least amount of arrests, then Troy Hill-Herrs Island, then far followed by Mt. Oliver Boro.

## Dataset #3
Since I'm in a group of two, this is part of the work I did for our third dataset.

```python
# importing contract rent data and looking at the first 10 values to get an idea of it
contract_rent_data = pd.read_csv("City of Pittsburgh Contract Rent.csv")

contract_rent_data.head(10)
```

```python
# Calculate the average estimate total for every neighborhood
average_estimates = contract_rent_data.groupby("Neighborhood")["Estimate; Total:"].mean()
average_estimates_sorted = average_estimates.sort_values(ascending=True)

# display values
average_estimates_sorted.head(10)
```

## Notes

* Chateu has the lowest contract rent average according to this data analysis. 

* According to the data set's website: "Aggregate contract rent is calculated by adding all of the contract rents for occupied housing units in an area. Aggregate contract rent is rounded to the nearest hundred dollars."

* So the reason Chateau showed up as 0 is because the estimated total, which we averaged, was lower than one hundred dollars.
